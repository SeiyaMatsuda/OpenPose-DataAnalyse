```python
import matplotlib.pyplot as plt
%matplotlib inline
import OpenPose_kaiseki as OP
```

```python
#時系列データ作成関数make_graf(人,pair)
def make_graf(h,p):
    #x軸
    x=[]
    #y軸
    y=[]
    #人数分のループ
    for r in range(0,len(OP.result_angle[0])):
        #x軸(フレーム数)のデータを格納
        x=x+[OP.result_angle[h][r][0]]
        #もし角度データがNoneの場合
        if OP.result_angle[h][r][p]==None:
            #y軸の値にNoneを追加
            if len(y)==0:
                y=y+[None]
            #y軸に1フレーム前の値を参照
            else: 
                y=y+[y[-1]]
        #角度データをy軸に追加
        else:
            y=y+[OP.result_angle[h][r][p]]
    #x軸，y軸を返す
    return x,y
#時系列データ作成に必要なx軸，y軸の取得
x,y1=make_graf(1,1)
x,y2=make_graf(1,2)
x,y3=make_graf(1,3)
x,y4=make_graf(1,4)
x,y5=make_graf(1,5)
x,y6=make_graf(1,6)
#以下プロット
plt.plot(x,y1,label="i=1")
plt.plot(x,y2,label="i=2")
plt.plot(x,y3,label="i=3")
plt.plot(x,y4,label="i=4")
plt.plot(x,y5,label="i=5")
plt.plot(x,y6,label="i=6")
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.)
plt.tight_layout()
#x軸のラベル指定
plt.xlabel("フレーム数",fontname="MS Gothic")
#y軸のラベル指定
plt.ylabel("角度",fontname="MS Gothic")
plt.show()
```
